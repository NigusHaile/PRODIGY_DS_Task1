```python
top_10_countries = df['Country Name'].value_counts().head(10)

# Plotting
bar_width = 0.9
x = range(len(top_10_countries.index))

plt.bar(top_10_countries.index, top_10_countries.values, width=bar_width)
plt.xlabel('Country Name')
plt.ylabel('Count')
plt.title('Distribution of Top 10 Countries')

plt.xticks(x, top_10_countries.index, rotation=45)
plt.tight_layout()
plt.show()
```
